```python
import pandas as pd
import random


output_file = 'output.csv'

df11 = pd.read_csv(output_file)

# Define the percentage of rows to remove 'pop' from (60% in this case)
percentage_to_remove = 0.6

# Create a function to randomly sample rows and remove 'pop'
def remove_pop(row):
    if 'pop' in row['actual_multi_genre']:
        genres = row['actual_multi_genre'].split(', ')
        if len(genres) > 1 and 'pop' in genres and random.random() < percentage_to_remove:
            genres.remove('pop')
            return ', '.join(genres)
    return row['actual_multi_genre']

# Apply the function to the 'actual_multi_genre' column
df11['actual_multi_genre'] = df11.apply(remove_pop, axis=1)

# Save the updated DataFrame to a new CSV file
df11.to_csv('output_removed_pop.csv', index=False)
```

```python
import pandas as pd
import random

# Replace 'output.csv' with the path to your CSV file
output_file = 'output.csv'

df13 = pd.read_csv(output_file)

# Define the percentage of "pop" instances to remove (60% in this case)
percentage_to_remove = 1.0

def remove_pop(row):
    genres = row['actual_multi_genre'].split(', ')
    new_genres = [genre for genre in genres if genre != 'pop' or (genre == 'pop' and random.random() > percentage_to_remove)]
    return ', '.join(new_genres)

df13['actual_multi_genre'] = df13.apply(remove_pop, axis=1)

# Save the updated DataFrame to a new CSV file
df13.to_csv('output_removed_pop.csv', index=False)




```
